```python
# RUSSIA

def russia(outlet):
    
    oasum = {"Vader_POS":0,"Vader_NEU":0,"Vader_NEG":0,
             "TB_POS": 0, "TB_NEU": 0, "TB_NEG":0
            }
    
    with open('Russia.csv', 'a') as fo2:
        csvWriter = csv.writer(fo2)
        csvWriter.writerow(['Article', 'V_Positive', 'V_Neutral', 'V_Negative', 'V_Compound', 'V_Sum_Pos', 'V_Sum_Neu', 'V_Sum_Neg', 'TB_Pos', 'TB_Neu', 'TB_Neg', 'TB_Polarity', 'TB_Subj'])
    
    for i in outlet:        
        with open(i, 'r', encoding = "utf-8") as fo:
            a = fo.read()

            from nltk.sentiment.vader import SentimentIntensityAnalyzer

            sid = SentimentIntensityAnalyzer()
            ss = sid.polarity_scores(a)
            summary = {"positive":0,"neutral":0,"negative":0}

            tb_summary = {"positive":0, "neutral":0, "negative": 0}
            analysis = TextBlob(a)

            # Conduct Vader Analysis
            if ss["compound"] == 0.0: 
                summary["neutral"] +=1
                oasum["Vader_NEU"] +=1
            elif ss["compound"] > 0.0:
                summary["positive"] +=1
                oasum["Vader_POS"] +=1
            else:
                summary["negative"] +=1
                oasum["Vader_NEG"] +=1

            # Conduct TextBlob Analysis    
            if analysis.polarity > 0.0:
                tb_summary["positive"] += 1
                oasum["TB_POS"] +=1
            elif analysis.polarity < 0.0:
                tb_summary["negative"] += 1
                oasum["TB_NEG"] +=1
            else:
                tb_summary["neutral"] += 1
                oasum["TB_NEU"] +=1
                
            with open('Russia.csv', 'a') as fo3:
                csvWriter = csv.writer(fo3)
                csvWriter.writerow([i, ss['pos'],ss['neu'],ss['neg'],ss['compound'], summary["positive"], summary["neutral"],summary["negative"], tb_summary["positive"], tb_summary["neutral"], tb_summary["negative"], analysis.polarity, analysis.subjectivity])        

    print(oasum)
    
    with open('RUS_OAS.csv', 'a') as fo5:
        csvWriter = csv.writer(fo5)
        csvWriter.writerow([['Vader_POS'],['Vader_NEU'],['Vader_NEG'],['TB_POS'],['TB_NEU'],['TB_NEG']])
        csvWriter.writerow([oasum['Vader_POS'],oasum['Vader_NEU'],oasum['Vader_NEG'],oasum['TB_POS'],oasum['TB_NEU'],oasum['TB_NEG']])
```
